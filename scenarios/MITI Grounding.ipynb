```python
import requests
import yaml
from IPython.display import Markdown, display
from tabulate import tabulate

import biolexica
```

```python
# This gets a pre-defined grounder
grounder = biolexica.load_grounder("phenotype")
```

```python
files = {
    "01-clinical",
    "02-biospecimine",
    "03-file",
    "04-channel",
    "05-ome-tiff-header",
    "06-preprocessed-data",
    "07-cell",
    "08-cell-state",
}

url = "https://raw.githubusercontent.com/miti-consortium/MITI/main/yaml/01-clinical.yaml"
data = yaml.safe_load(requests.get(url, timeout=3).text)
```

```python
for key, key_data in data.items():
    values = key_data.get("valid-values")
    if not values:
        continue

    display(Markdown(f"### {key}"))
    rows = []
    for value in values:
        match = grounder.get_best_match(value)
        if match is None:
            rows.append((value, None, None))
        else:
            rows.append(
                (
                    value,
                    f"[{match.curie}](https://bioregistry.io/{match.curie})",
                    match.name,
                )
            )

    display(
        Markdown(tabulate(rows, headers=["text", "grounding", "standard name"], tablefmt="github"))
    )
```
